```python
# Load the Flight Prices dataset in chunks
file_path = '/content/itineraries.csv'  # Adjust if necessary
chunk_size = 15
output = []

# Process the dataset in chunks
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    # Define categories
    categories = ['legId', 'startingAirport', 'travelDuration', 'destinationAirport']  # Adjust based on actual columns
    details = chunk[categories]
    details['count'] = 1
    summary = details.groupby(categories).sum().reset_index()
    output.append(summary)  # Collect each chunk summary

# Concatenate all collected dataframes into a single dataframe
final_output = pd.concat(output, ignore_index=True)

# Final aggregation across all chunks
final_output = final_output.groupby(categories).sum().reset_index()
final_output.to_csv('aggregated-flight-prices.csv', index=False)
```
